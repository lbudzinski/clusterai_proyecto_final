## Limpiar Data

```python
# Voy a crear un el campo llamado 'plazo_resoluci√≥n' = diferencia entre la fechas de apertura y cierre de contacto en DIAS
# Esta FEATURE voy a usarla para ver el tiempo medio de resolucion de los tramites y un boxplot de los mismos.

#2017
#NO tiene la FEATURE fecha_cierre_contacto... asi que no podremos visualizar el plazo de resolucion

#2018
suaci_2018_clean['fecha_cierre_contacto'] = pd.to_datetime(suaci_2018_clean['fecha_cierre_contacto'])
suaci_2018_clean['fecha_ingreso'] = pd.to_datetime(suaci_2018_clean['fecha_ingreso'])
suaci_2018_clean['plazo_resolucion'] = suaci_2018_clean['fecha_cierre_contacto'] - suaci_2018_clean['fecha_ingreso']
suaci_2018_clean['plazo_resolucion'] = suaci_2018_clean['plazo_resolucion'].dt.days
#Tiempo Medio de resolucion 2018
suaci_2018_clean_plazos_categoria_media = suaci_2018_clean.groupby('categoria')['plazo_resolucion'].mean()


#2019
suaci_2019_clean['fecha_cierre_contacto'] = pd.to_datetime(suaci_2019_clean['fecha_cierre_contacto'])
suaci_2019_clean['fecha_ingreso'] = pd.to_datetime(suaci_2019_clean['fecha_ingreso'])
suaci_2019_clean['plazo_resolucion'] = suaci_2019_clean['fecha_cierre_contacto'] - suaci_2019_clean['fecha_ingreso']
suaci_2019_clean['plazo_resolucion'] = suaci_2019_clean['plazo_resolucion'].dt.days
#Tiempo Medio de resolucion 2019
suaci_2019_clean_plazos_categoria_media = suaci_2019_clean.groupby('categoria')['plazo_resolucion'].mean()
```
